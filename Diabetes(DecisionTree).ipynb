```python
ind=dia[['age','mass','insu','plas']]#independent
dep=dia['class']#dependent
```

```python
LR=LogisticRegression()
```

```python
LR.fit(ind,dep)
```

```python
age=int(input("Enter the age: "))
mass=int(input("Enter the mass: "))
insulin=int(input("Enter the insulin level: "))
plasma=int(input("Enter the plasma level: "))
pred= LR.predict([[age,mass,insulin,plasma]])
print(pred)
```

```python
LR.score(ind,dep)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

### Decision Tree Model

```python
DT = DecisionTreeClassifier()
DT.fit(ind, dep)
print(DT)
```

#### Predicting with Decision Tree

```python
age_dt=int(input("Enter the age: "))
mass_dt=int(input("Enter the mass: "))
insulin_dt=int(input("Enter the insulin level: "))
plasma_dt=int(input("Enter the plasma level: "))

pred_dt = DT.predict([[age_dt, mass_dt, insulin_dt, plasma_dt]])
print(f"Decision Tree prediction: {pred_dt}")
```

#### Decision Tree Model Score

```python
print(f"Decision Tree score: {DT.score(ind, dep)}")
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

### Decision Tree Graph

```python
plt.figure(figsize=(20,15))
plot_tree(DT, filled=True, feature_names=ind.columns.tolist(), class_names=DT.classes_.tolist())
plt.title("Decision Tree Visualization")
plt.show()
```
